```python
import pandas as pd
import numpy as np
from datetime import datetime
from pandas import Series
```

# Import The Data

We crossed information between two data sources. We took lists of top 100 songs weekly from Bilboard and matched it with the features from spotify. The data contains top 100 lists from 1990 until 2021.

```python
df = pd.read_csv("./data/TimeSeries.csv")
```

```python
df.head()
```

```python
df.reset_index(drop=True)
```

### Convert date column to datetime format

```python
count = 0
for time in df["date"]:
    try:
        pd.to_datetime(time, format='%m/%d/%Y')
    
    except ValueError:
        print("Incorrect data format ", time)
        count = count + 1
```

```python
count
```

```python
df['date'] = pd.to_datetime(df['date'], format='%m/%d/%Y')
df['date']
```

### Add the month of the year for each week

```python
df['month_year'] = pd.to_datetime(df['date']).dt.to_period('M')
```

```python
df
```

### Normalizing the data
Making sure all features are between 0 and 1

```python
for col in ['loudness',"tempo","key"]:
    col_min = df[col].min()
    col_max = df[col].max()
    df[col] = ((df[col] - col_min) / (col_max - col_min))
```

```python
df
```

```python
df.to_csv("./data/DF_Daily_Time_Series.csv", index=False)
```
